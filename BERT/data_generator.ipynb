```python
df1= df.drop(['District','PS Name','FIR No', 'Date Report', 'Date Accident', 'Time Report', 'Time Accident',  'Sections',  'Death',
       'Grievous', 'Minor', 'Pedestrian', 'Cyclist', 'Latitude', 'Longitude','Type Area', 'City/Town/ Village', 'T -Junction', ],axis=1)
display(df1)
```

```python
list_of_columns = df1.columns
list_of_columns
```

```python
# Dictionary to store unique values
unique_values = {}

# Loop through columns and get unique values
for col in list_of_columns:
    unique_values[col] = df[col].unique().tolist()

# Print unique values
for col, values in unique_values.items():
    print(f"{col}: {values}")
```
